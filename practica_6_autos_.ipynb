```python
df = pd.read_csv("autos2.csv")
n = len(df["price"])
K = int( 1 + math.log2(n))
print("valor de n=", n ,"\nNUmeros de intervalos=", K)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['price'], bins=K, edgecolor='g', alpha=0.7)
plt.xlabel('Precios de los automóviles')
plt.ylabel('Frecuencia')
plt.title('Histograma de precios de automóviles')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='body-style', y='price', data=df)
plt.xlabel('Tipos de vehículos')
plt.ylabel('Precios de los automóviles')
plt.title('Precio vs Tipo de Vehículo')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='drive-wheels', y='price', data=df)
plt.xlabel('Tracción del vehículo')
plt.ylabel('Precios de los automóviles')
plt.title('Precio vs Tracción del Vehículo')
plt.show()
```

```python

plt.figure(figsize=(10, 5))
sns.boxplot(x='engine-location', y='price', data=df)
plt.xlabel('Ubicación del motor')
plt.ylabel('Precios de los automóviles')
plt.title('Precio vs Ubicación del Motor')
plt.show()
```
