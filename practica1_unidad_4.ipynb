```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
media = 70
desv_std = 4
num_datos = 500


datos = np.random.normal(media, desv_std, num_datos)
datos[0:50]
```

```python

x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)


pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desv_std **2))
```

```python

pdf.size, num_datos
```

```python

plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribuci√≥n Normal ')
plt.xlabel('Peso en Kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python


plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media,2), np.round(desv_std, 2)))
plt.title('Funci√≥n de Densidad de Probabilidad (PDF) - Distrubuci√≥n Normal')
plt.xlabel('Peso en Kg de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot( x, pdf, 'k', linewidth=2, )
plt.title('Distribuci√≥n Normal ')
plt.xlabel('Peso en Kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(6, 4))


media = 70
desv_std = 4
total_estudiantes = 500

peso_limite = 75


prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)
```

```python
estudiantes_menos_75kg = prob_menos_75kg * total_estudiantes

print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("N√∫mero de estudiantes que pesan menos de 75 kg:", estudiantes_menos_75kg)

# Graficar la distribuci√≥n normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.xlabel('Peso en Kg de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.title('Probabilidad de que un estudiante pese menos de 75 kg, \n n=70 p=4 x=500')
plt.grid(True)
plt.show()
```

```python
peso_min = 68
peso_max = 72


x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)


prob_entre = norm.cdf(peso_max, media, desv_std) - norm.cdf(peso_min, media, desv_std)
estudiantes_entre = prob_entre * total_estudiantes


plt.figure(figsize=(6, 4))
plt.plot(x, y, 'k', linewidth=2, label='Distribuci√≥n Normal')
plt.fill_between(x, y, where=(x >= peso_min) & (x <= peso_max), color='skyblue', alpha=0.5,
                 label='Entre 68 y 72 kg')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribuci√≥n Normal: Estudiantes entre 68 y 72 kg')
plt.legend()
plt.grid(True)
plt.show()

print("Probabilidad entre 68 y 72 kg:", prob_entre)
print("N√∫mero aproximado de estudiantes entre 68 y 72 kg:", estudiantes_entre)
```
