```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
```

```python
grupo1 = pd.read_csv('grupo1.csv')
grupo1.head()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(grupo1['Calificacion'], bins=10, kde=True, color='skyblue')
plt.title('Histograma de Calificaciones')
plt.xlabel('Calificacin Final')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
X = grupo1[['HorasEstudio']]
y = grupo1['Calificacion']

modelo = LinearRegression()
modelo.fit(X, y)

pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print(f"\nEcuacin de la recta: y = {pendiente:.2f}x + {intercepto:.2f}")
coef_pearson, p_valor = pearsonr(grupo1['HorasEstudio'], grupo1['Calificacion'])
print(f"\nCoeficiente de correlacin de Pearson: {coef_pearson:.4f}")
print(f"Valor p asociado: {p_valor:.4f}")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=grupo1, color='blue', label='Datos')
plt.plot(grupo1['HorasEstudio'], modelo.predict(X), color='red', label='Lnea de Regresin')
plt.title('HorasEstudio vs Calificacion')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacin Final')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

1. Cul es la interpretacin del coeficiente de regresin en trminos de la relacin entre el tiempo de estudio y las calificaciones?
El coeficiente de regresin pendiente indica cunto se espera que cambie la calificacin de un estudiante por cada hora adicional de estudio. La pendiente es 2.47 puntos.
2. Cmo se interpreta el coeficiente de correlacin en trminos de fuerza y direccin?
El coeficiente de correlacin de Pearson (r) mide la fuerza y direccin de una relacin lineal entre dos variables.
El valor de la correlacin es 0.95
3. El modelo de regresin lineal es estadsticamente significativo?
Para saberlo, se analiza el valor p asociado al coeficiente de correlacin el valor fue El valor fue 0.91
4. Qu tan confiable es el modelo de regresin para predecir las calificaciones finales?
es muy confiable
