```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
```

```python
grupo1 = pd.read_csv('grupo1.csv')
grupo1.head()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(grupo1['Calificacion'], bins=10, kde=True, color='skyblue')
plt.title('Histograma de Calificaciones')
plt.xlabel('Calificación Final')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
X = grupo1[['HorasEstudio']]
y = grupo1['Calificacion']

modelo = LinearRegression()
modelo.fit(X, y)

pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print(f"\nEcuación de la recta: y = {pendiente:.2f}x + {intercepto:.2f}")
coef_pearson, p_valor = pearsonr(grupo1['HorasEstudio'], grupo1['Calificacion'])
print(f"\nCoeficiente de correlación de Pearson: {coef_pearson:.4f}")
print(f"Valor p asociado: {p_valor:.4f}")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=grupo1, color='blue', label='Datos')
plt.plot(grupo1['HorasEstudio'], modelo.predict(X), color='red', label='Línea de Regresión')
plt.title('HorasEstudio vs Calificacion')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación Final')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

1. ¿Cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones?
El coeficiente de regresión pendiente indica cuánto se espera que cambie la calificación de un estudiante por cada hora adicional de estudio. La pendiente es 2.47 puntos.
2. ¿Cómo se interpreta el coeficiente de correlación en términos de fuerza y dirección?
El coeficiente de correlación de Pearson (r) mide la fuerza y dirección de una relación lineal entre dos variables.
El valor de la correlación es 0.95
3. ¿El modelo de regresión lineal es estadísticamente significativo?
Para saberlo, se analiza el valor p asociado al coeficiente de correlación el valor fue El valor fue 0.91
4. ¿Qué tan confiable es el modelo de regresión para predecir las calificaciones finales?
es muy confiable
